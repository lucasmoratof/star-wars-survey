# Convert a column into from to Boolean:

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
yes_no = {'Yes': True, 'No': False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
```

```python
star_wars.head()
```

Doing the same thing for the columns asking which movies the person have seen from the franchise, and renaming the column to make them more intuitive.

```python
import numpy as np
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

Now let's convert to .astype(float) the columns which evalute how much people like each movie, then rename those columns to make it easy to understand.

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
rank_names = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
             'Unnamed: 10': 'ranking_2',
             'Unnamed: 11': 'ranking_3',
             'Unnamed: 12': 'ranking_4',
             'Unnamed: 13': 'ranking_5',
             'Unnamed: 14': 'ranking_6'
            }
star_wars.rename(columns=rank_names, inplace=True)
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
% matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

# Rankings 
On the bar chart above we can see the mean of the votes on each of Star Wars movie.<br>
The metric is confusing, as people needed to choose between 1 and 6, being 1 the highest score and 6 the lowest.

# View counts
It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

# Gender Analysis

```python
plt.subplot(1,2,1)
males[males.columns[3:9]].sum().plot(kind='bar', title='View Count: Males ')

plt.subplot(1,2,2)
males[males.columns[9:15]].mean().plot(kind='bar', title='Star Wars Ranking: Males')
```

```python
plt.subplot(1,2,1)
females[females.columns[3:9]].sum().plot(kind='bar', title='View Count: Females ')

plt.subplot(1,2,2)
females[females.columns[9:15]].mean().plot(kind='bar', title='Star Wars Ranking: Females')
```

We can observe:
- More males watched the first movies;
- There is a tendance of preference for the new movies on the rank.

# Movies ranking by education

```python
star_wars.Education.value_counts()
```

```python
# using knowledege and logic from data school! movies[movies.genre.isin(['Crime','Drama','Action'])]

group_1 = star_wars[star_wars['Education'].isin(['Graduate degree','Bachelor degree'])]
group_2 = star_wars[star_wars['Education'].isin(['Some college or Associate degree','High school degree','Less than high school degree'])]
```

```python
plt.subplot(1,2,1)
group_1[group_1.columns[9:15]].mean().plot(kind='bar',title='Higher Educational Level')
plt.subplot(1,2,2)
group_2[group_2.columns[9:15]].mean().plot(kind='bar',title='Lower Educational Level')
```

There is not a significant difference on the movies rating based on the educational level.

# Cleaning columns, converting str to float to perform analysis:

```python
star_wars['Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.'].value_counts(dropna=False)
```

```python
map_char = {
    'Very favorably': 6,
    'Somewhat favorably': 5,
    'Neither favorably nor unfavorably (neutral)': 4,
    'Unfamiliar (N/A)': 3,
    'Somewhat unfavorably': 2,
    'Very unfavorably': 1
}

for col in star_wars.columns[15:29]:
    star_wars[col] = star_wars[col].map(map_char)

star_wars[star_wars.columns[15:29]] = star_wars[star_wars.columns[15:29]].astype(float)    
star_wars.head()
```

```python
star_wars[star_wars.columns[15:29]].mean().plot(kind='bar')
```

We created a simple bar plot with the characters and the mean of evaluation of each one. If we discover the name, it is possible to say the favourite (on the average) of the public.
